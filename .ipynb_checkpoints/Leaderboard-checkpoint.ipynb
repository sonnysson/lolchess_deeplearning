```python
from bs4 import BeautifulSoup
import urllib.request
import ssl
import os
import time

context = ssl._create_unverified_context()

def Requesting_url():
    t = 0
    e = 100
    while(t < 10):
        try:
            webpage = urllib.request.urlopen("https://lolchess.gg/leaderboards", context=context)
        except urllib.error.URLError as e:
            t += 1
            time.sleep(0.5)
            pass
        if(e < 300):
            break
        else:
            print("%d"%e+" Error")
    return webpage

webpage = Requesting_url()
```

```python
soup = BeautifulSoup(webpage, "html.parser")
print(soup)
```
