```python
import numpy as np
import math
import matplotlib.pyplot as plt

t=np.linspace(1,100,1000)
x1=5*np.sin(t/(2*np.pi))
x2=5*np.cos(t/(0.5*np.pi))+50
x3=5*np.sin(t/(1*np.pi))+100

plt.subplot(3,1,1)
plt.plot(t,x1,color='blue')
plt.title('x1')
plt.ylabel('y')

plt.subplot(3,1,2)
plt.plot(t,x2,color='red')
plt.title('x2')
plt.ylabel('y')

plt.subplot(3,1,3)
plt.plot(t,x3,color='green')
plt.title('x3')
plt.ylabel('y')


plt.tight_layout()
plt.show()

plus=100
start = 0
end = plus

while end<=1000:
    plt.scatter(np.mean(x1[start:end]),np.var(x1[start:end]),color="blue")
    plt.scatter(np.mean(x2[start:end]),np.var(x2[start:end]),color="red")
    plt.scatter(np.mean(x3[start:end]),np.var(x3[start:end]),color="green")
    start += plus
    end += plus
    
plt.show()

# plt.plot(t,np.var(x1),color="red")
# plt.title("Feature Space")
# plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
# x 데이터는 행 ,열 reshape() ones_array랑 합치기위해 벡터 동일화
# 차원수 ,벡터 같은지 항상 확인할것!!
# numpy.reshape(Calumn,Row)

x = np.array([0.1,0.4,0.7,1.2,1.3,1.7,2.2,2.8,3.0,4.0,4.3,4.4,4.9]).reshape(-1,1)
y = np.array([0.5,0.9,1.1,1.5,1.5,2.0,2.2,2.8,2.7,3.0,3.5,3.7,3.9])

# 1도 된 array 생성 (13행 1열)
# np.info(x) 로 행,열 개수 확인
ones_array = np.ones(shape=(13,1),dtype=np.int8)

# concatenate((array,array1),axis = 0 or 1 # default = 0 (행으로 합차기),1 열로 합치기 ) - array 합치기
x_theta = np.concatenate((ones_array,x),axis=1)

# x 세타 스타 계산
# numpy.dot = 곱셈 
# numpy.linalg.inv = 역행렬 (-1)
# numpy.transpose = 전치행렬 (T)
theta_star = np.dot(np.dot(np.linalg.inv(np.dot(np.transpose(x_theta),x_theta)),np.transpose(x_theta)),y).reshape(-1,1)

# 방법이 두가지
t = np.linspace(0,6,100)
t = np.arange(0,7,0.01)

# 진짜 회귀 분석값
y_predict = np.dot(x_theta,theta_star)
# 내가 예측한 값
my_predict = theta_star[0]*t +theta_star[1]

plt.title("Regression",fontsize = 15)
plt.subplot()
plt.xlabel("X",fontsize = 10)
plt.ylabel("Y",fontsize = 10)
plt.plot(x,y,"ko",label="data")
plt.plot(x,y_predict,color="red",label='y_predict')
plt.plot(t,my_predict,color="blue",label='my_predict')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.1, 0.4, 0.7, 1.2, 1.3, 1.7, 2.2, 2.8, 3.0, 4.0, 4.3, 4.4, 4.9]).reshape(-1,1)
y = np.array([0.5, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.8, 2.7, 3.0, 3.5, 3.7, 3.9]).reshape(-1,1)

plt.plot(x,y,"ko",label="data")

m = x.shape[0]
A = np.hstack([np.ones([m,1]),x])
A = np.asmatrix(A)

theta = (A.T*A).I*A.T*y
y_predict = A*theta
plt.plot(x,y_predict,color="red")
```
