## 2. Thinking humanly ( 사람처럼 생각
    인지과학 (Cognitive Science) : 사람의 마음/생각에 대해서 실증 가능한 이론을 정립 -> 구현
    내성법 : 스스로에 대해 생각해보는것
    심리학적 실험: 사람이 어떻게 생각하는지?
    
## 3. Thinking Rationally ( 생각의 법칙 기반 접근
    생각의 절차를 모형화 하여 주어진 사실로 결론을 유추
    Notation(표기) and Rules(규칙 : 삼단논법)  를 따른다
    개념적인 것과 실제의 차이가 있다 (AI winter)
    
## 4. Acting Rationally : the rational agent approach( 합리적 에이전트
    - 주어진 정보에 기반하여 목표 달성의 가능성을 가장 높일수 있는것으로 기대되는 절차를 수행하는 시스템
    - 행위가 있는것들만 에이전트라고 부른다.
    - Rational Agent(합리적 에이전트)
